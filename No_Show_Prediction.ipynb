```python
show['PatientId']=show['PatientId'].astype(int)
```

```python
show.head()
```

```python
show.drop('PatientId',axis=1,inplace=True)
```

```python
show.drop('AppointmentDay',axis=1,inplace=True)
```

```python
show['Gender']=show['Gender'].map({'M':0,'F':1})
show['No-show']=show['No-show'].map({'No':0,'Yes':1})
```

```python
show
```

```python
show['ScheduledDate']=show['ScheduledDay'].str.split("T").str[0]
show['ScheduledTime']=show['ScheduledDay'].str.split("T").str[1]
show.drop('ScheduledDay',axis=1,inplace=True)
```

```python
show
```

```python
show=show.rename(columns={'Hipertension':'Hypertension'})
new=['AppointmentID','ScheduledDate','ScheduledTime','Gender','Age','Neighbourhood','Scholarship','Hypertension','Diabetes','Alcoholism','Handcap','SMS_received','No-show']
show=show[new]
show
```

```python
x=show[['Hypertension','Diabetes','Alcoholism']].values
y=show['No-show'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.20)
```

```python
x_train=x_train.astype(float)
x_test=x_test.astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
show_model=LogisticRegression()
```

```python
show_model.fit(x_train,y_train)
```

```python
y_pred=show_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
