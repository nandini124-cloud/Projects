```python
x=df[['Time_spent_Alone','Social_event_attendance','Post_frequency']].values
y=df['Personality'].values
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train=x_train.astype(float)
x_test=x_test.astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
classifier=SVC( kernel='rbf',random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred1=classifier.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred1)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier1=KNeighborsClassifier(n_neighbors=5)
classifier1.fit(x_train,y_train)
```

```python
y_pred2=classifier1.predict(x_test)
```

```python
cn=confusion_matrix(y_test,y_pred2)
cn
```

```python
sns.heatmap(cn,annot=True)
```

```python
acc=accuracy_score(y_test,y_pred2)
```

```python
acc
```
